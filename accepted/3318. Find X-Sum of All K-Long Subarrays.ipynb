```python
from collections import Counter
from typing import List

def x_sum_subarrays(nums: List[int], k: int, x: int) -> List[int]:
    n = len(nums)
    answer = []
    
    for i in range(n - k + 1):
        window = nums[i:i + k]
        freq = Counter(window)
        
        items = list(freq.items())
        items.sort(key=lambda item: (-item[1], -item[0]))
        
        total = 0
        take = min(x, len(items))
        for val, count in items[:take]:
            total += val * count
        
        answer.append(total)
    
    return answer
```
